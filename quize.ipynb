```python
def quit():
    print("\nThankuu!!!")
    sys.exit()
```

```python
df = pd.read_csv('ques.csv')
df.head()
```

```python
count=0
def start():
    global count
    x=random.randint(0,40)
    print(df['Questions'][x])
    print(df['optionA'][x])
    print(df['optionB'][x])
    print(df['optionC'][x])
    print(df['optionD'][x])
    ans=input('Enter your choice.. ')
    if ans==df['Answer'][x]:
        print("\nCorrect answer!!\n\n")
        count=count+1
        if count<=10:
            start()
        else:
            print("\n\n \t\t**************** CONGRATULATION ****************")
            print("\n\t\tYou won Rs 100000.00")
            print("\t\t Thank You!!")
            print("Do you want to play again??\nIf yes press 'Y' otherwise 'N' ")
            result=input('Enter your choice.. ')
            if result =='Y'or result=='y':
                count=0
                start()
            else:
                quit()
            
    else:
        print("\nYou lost!!")
        print("Correct answer is  {}".format(df['Answer'][x]))
        score=count*100000
        print("Better luck next time!!!")
        print("\n\t\t**************** CONGRATULATION *****************")
        print("\nYou won Rs {}".format(score))
        print("\nDo you want to play again??\nIf yes press 'Y' otherwise 'N' ")
        result=input('Enter your choice.. ')
        if result =='Y' or result=='y':
            print("start again")
            count=0
            start()
        else:
            quit()
```

```python

def main_func():
    homepage()
    choice=input('enter your choice.. ')
    if choice=='S' or choice=='s':
        start()
        main_func()
    elif choice=='H' or choice=='h':
        #sys("cls")
        Help()
        main_func()
    elif choice=='Q' or choice=='q':
        quit()
    else:
        print('wrong choice!!')
        main_func()
```
